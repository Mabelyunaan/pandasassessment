```python
if 'price' in df.columns:
    df['price']=df['price'].astype(str).replace('$,', '', regex = True).astype(float)
    #replaces the price column to strictly numericals to allow mathematical expression and fills th missing values with 0
    
```

```python
if 'last_review' in df.columns:
    df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce') 
    #checks if there's a column named last_review and if tehre isnt the code won't run , 
    #if there is the pd.to_datetime dectects all the date format and error ='coerce' 
   #if the date can't be parsed turns into NaT instead of crashing
   #we can now perform time based analysis
```

```python
if 'reviews_per_month' in df.columns:
    df['reviews_per_month'] = df['reviews_per_month'].fillna('unknown')
    #fills missing values in the reviews per month column with unkown
    
```

```python
if 'neigbourhood_group' in df.columns:
    df['neigbourhood_group'] = df['neigbourhood_group'].fillna('unknown')
    #fills missing values with unkown
```

```python
if 'price' in df.columns:
    df =df[df['price'] > 0]
```

```python
if 'availability_365' in df.columns:
    df = df[df['availability_365'] > 0]

print("cleaned dataset:" , df.shape)
```

```python
#data enrichment
```

```python
# ---------------------------------------------
# 🧩 3. DATA ENRICHMENT
# ---------------------------------------------

# Create price_per_booking = price * minimum_nights
if all(col in df.columns for col in ['price', 'minimum_nights']):
    df['price_per_booking'] = df['price'] * df['minimum_nights']

# Bucket availability into categories
def categorize_availability(days):
    if days > 300:
        return 'Full-time'
    elif days >= 100:
        return 'Part-time'
    else:
        return 'Rare'

if 'availability_365' in df.columns:
    df['availability_category'] = df['availability_365'].apply(categorize_availability)

df[['availability_365', 'availability_category']].head()
```

```python
# data analysis
if 'neighbourhood' in df.columns:
    top10_neigh = (
        df.groupby('neighbourhood')['price']
        .mean()
        .sort_values(ascending=False)
        .head(10)
    )
    print("\nTop 10 Most Expensive Neighborhoods:\n", top10_neigh)

 #Average availability and price by room type
if 'room_type' in df.columns:
    room_summary = df.groupby('room_type')[['availability_365', 'price']].mean()
    print("\nAverage Availability & Price by Room Type:\n", room_summary)

#Host with the most listings
if 'host_id' in df.columns:
    top_host = (
        df.groupby(['host_id', 'host_name'])
        .size()
        .reset_index(name='listing_count')
        .sort_values(by='listing_count', ascending=False)
        .head(1)
    )
    print("\nHost with Most Listings:\n", top_host)

#Average price by neighbourhood_group (borough)
if 'neighbourhood_group' in df.columns:
    borough_price = df.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)
    print("\nAverage Price by Borough/District:\n", borough_price)

# Listings that have never been reviewed
if 'number_of_reviews' in df.columns:
    never_reviewed = df[df['number_of_reviews'] == 0].shape[0]
    print(f"\nNumber of listings never reviewed: {never_reviewed}")
```
